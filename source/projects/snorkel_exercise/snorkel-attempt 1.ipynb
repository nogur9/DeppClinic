```python

# Label notes mentioning urgent risk as RISK
@labeling_function()
def lf_urgent_risk(note_text):
    keywords = ["חרדות", "טיפול פסיכולוגי דחוף", "מתקש", "עצבות"]
    if any(keyword in note_text for keyword in keywords):
        return "RISK"
    return "ABSENT"

# Label notes expressing concern as RISK or NON-RISK based on context
@labeling_function()
def lf_concern(note_text):
    if "מודאגת" in note_text:
        return "RISK"
    return "ABSENT"

# Label notes mentioning medication treatment as RISK or NON-RISK based on urgency
@labeling_function()
def lf_medication(note_text):
    if "תרופתי בלבד" in note_text:
        return "RISK"
    elif "עדיין לא התחילה לקחת" in note_text:
        return "RISK"
    return "ABSENT"
```
