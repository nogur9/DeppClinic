```python
from itertools import combinations

def iterate_over_feature(full_feature_list, max_features=None):
    """
    Generate all possible combinations of features from the full_feature_list.
    
    Parameters:
    - full_feature_list (list): List of all available features.
    - max_features (int): Maximum number of features to include in any combination.
      If None, use all features.
    
    Yields:
    - tuple: A tuple representing a subset of features.
    """
    if max_features is None:
        max_features = len(full_feature_list)

    for r in range(1, max_features + 1):
        for combo in combinations(full_feature_list, r):
            yield combo


def filter_features(X, feats):
    return X[feats]
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


def filter_features(X, feats):
    return X[list(feats)]

for feats in iterate_over_feature(QUESTIONS):
    print(feats)
    X_train, X_test, y_train, y_test=load_and_split_data()
    X_train = filter_features(X_train, feats)
    X_test = filter_features(X_test, feats)
    model = CatBoostClassifier(verbose=False, auto_class_weights='Balanced')
    model.fit(X_train, y_train)
    evaluate_model(model, X_test, y_test)
```
